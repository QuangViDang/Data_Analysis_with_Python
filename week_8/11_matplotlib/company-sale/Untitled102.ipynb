# Visualizing Company Sale data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/matplotlib/company_sales_data.csv).

```python
df = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv')
df.head(5)
```

### Exercise 1: Read Total profit of all months and show it using a line plot

![img](img/01.png)

```python
profit_list = df['total_profit'].tolist()
month_list = df['month_number'].tolist()
plt.plot(month_list , profit_list, label ='Company profit per month')
plt.legend(loc = 'upper left')
plt.xticks(month_list)
plt.yticks([100000,200000,300000,400000,500000])
plt.xlabel('Month number')
plt.ylabel('Profit')
plt.show()
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties

![img](img/02.png)

```python
profit_list = df['total_profit'].tolist()
month_list = df['month_number'].tolist()
plt.plot(month_list , profit_list, label ='Company sales data of last year', c = 'r', marker = 'o' ,markerfacecolor='c', linestyle=':', linewidth=5)
plt.legend(loc ='upper right')
plt.xticks(month_list)
plt.yticks([0,100000,200000,300000,400000,500000])
plt.xlabel('Month number')
plt.ylabel('Profit')
plt.show()
```

### Exercise 3: Read all product sales data and show it  using a multiline plot

![img](img/03.png)

```python
month_list = df['month_number'].tolist()
facecream_list = df['facecream'].tolist()
facewash_list = df['facewash'].tolist()
toothpaste_list = df['toothpaste'].tolist()
bathingsoap_list = df['bathingsoap'].tolist()
shampoo_list = df['shampoo'].tolist()
moisturizer_list = df['moisturizer'].tolist()

plt.plot(month_list, facecream_list,   label = 'Face cream Sales Data', marker='o', linewidth=3)
plt.plot(month_list, facewash_list,   label = 'Face Wash Sales Data',  marker='o', linewidth=3)
plt.plot(month_list, toothpaste_list, label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(month_list, bathingsoap_list, label = 'BathingSoap Sales Data', marker='o', linewidth=3)
plt.plot(month_list, shampoo_list, label = 'Sampoo Sales Data', marker='o', linewidth=3)
plt.plot(month_list, moisturizer_list, label = 'Moisturizer Sales Data', marker='o', linewidth=3)

plt.legend(loc ='upper left')
plt.xticks(month_list)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.xlabel('Month number')
plt.ylabel('Profit')
plt.title('hehe')
plt.show()
```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
![img](img/04.png)

```python
month_list = df['month_number'].tolist()
toothpaste_list = df['toothpaste'].tolist()
plt.scatter(month_list, toothpaste_list, label = 'Face cream Sales Data')
plt.legend(loc = 'upper left')
plt.xticks(month_list)
plt.yticks([4500,5000,5500,6000,6500,7000,7500,8000])
plt.grid(True,color = 'green', linewidth= 1, linestyle=":")
plt.xlabel('Month Number')
plt.ylabel('Number of units Sold')
plt.show()
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

![img](img/05.png)

```python
import numpy as np
month_list = df['month_number'].tolist()
facecream_list = df['facecream'].tolist()
facewash_list = df['facewash'].tolist()

x = np.arange(len(month_list))
width = 0.35

plt.bar(x - width/2, facecream_list, width, label = 'Face Cream Sales Data')
plt.bar(x + width/2, facewash_list , width, label = 'Face Wash Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc = 'upper left')
plt.xticks(month_list)
plt.yticks([500,1000,1500,2000,2500,3000,3500])
plt.grid(True,color = 'green', linewidth= 1, linestyle="--")
plt.show()
```

### Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

![img](img/06.png)

```python
month_list = df['month_number'].tolist()
bathingsoap_list = df['bathingsoap'].tolist()
plt.bar(month_list , bathingsoap_list)
plt.xticks(month_list)
plt.yticks([2000,4000,6000,8000,10000,12000,14000])

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('bathingsoap sales data')
plt.show()
```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

![img](img/07.png)

```python
profitList = df ['total_profit'].tolist()
profit_range = [150000, 200000, 250000, 300000, 350000]
plt.hist(profitList,profit_range, label = 'Profit data')

plt.legend(loc='upper left')
plt.xticks(profit_range)

plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.title('Profit data')
plt.show()
```

### Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

![img](img/08.png)

```python
labels = ['FaceWash', 'FaceCream', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
salesData = [df['facewash'].sum(),df['facecream'].sum(),df['toothpaste'].sum(), df['bathingsoap'].sum(), df['shampoo'].sum(), df['moisturizer'].sum()]
plt.pie(salesData, labels = labels)
plt.legend(loc = 'lower right')
plt.axis('equal')

plt.title('Sales data')
plt.show()
```

### Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

![img](img/09.png)

```python
month_list = df['month_number'].tolist()
bathingsoap_list = df['bathingsoap'].tolist()
facewash_list = df['facewash'].tolist()

fig2 = plt.figure(figsize = (10,5))
ax1, ax2 =  fig2.subplots(2,1)

ax1.plot(month_list, bathingsoap_list , c = 'k', marker = 'o', linewidth = 3 )
ax1.set_title('Sales data of a Bathingsoap')

ax2.plot(month_list, facewash_list , c = 'r', marker = 'o', linewidth = 3)
ax2.set_title('Sales data of a facewash')

plt.xticks(month_list)
plt.xlabel('Month number')
plt.ylabel('Sales units  in number')
plt.show()
```

### Exercise 10: Read all product sales data and show it using the stack plot

![img](img/10.png)

```python
month_list = df['month_number'].tolist()
facecream_list = df['facecream'].tolist()
facewash_list = df['facewash'].tolist()
toothpaste_list = df['toothpaste'].tolist()
bathingsoap_list = df['bathingsoap'].tolist()
shampoo_list = df['shampoo'].tolist()
moisturizer_list = df['moisturizer'].tolist()

plt.plot([], [], c = 'm'  ,label = 'Face cream Sales Data', linewidth=5)
plt.plot([], [], c = 'c'  ,label = 'Face Wash Sales Data', linewidth=5)
plt.plot([], [], c = 'r' ,label = 'ToothPaste Sales Data', linewidth=5)
plt.plot([], [], c = 'k', label = 'BathingSoap Sales Data', linewidth=5)
plt.plot([], [], c = 'g' , label = 'Sampoo Sales Data', linewidth=5)
plt.plot([], [], c = 'y' ,label = 'Moisturizer Sales Data',  linewidth=5)
plt.legend(loc ='upper left')

plt.stackplot(month_list, facecream_list, facewash_list, toothpaste_list, bathingsoap_list, shampoo_list, moisturizer_list, colors=['m','c','r','k','g','y'])

plt.xticks(month_list)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])

plt.xlabel('Month number')
plt.ylabel('Profit')
plt.title('AIII product sales data using stack plot')
plt.show()
```
