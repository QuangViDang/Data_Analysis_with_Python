### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv).

```python
import urllib.request

url = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv' # URL của tập tin cần đọc
data = pd.read_table(url)   # Đọc dữ liệu tại URL
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = data
```

### Step 4. See the first 10 entries

```python
chipo.head(10)
```

### Step 5. Create a histogram of the top 5 items bought

```python
import collections as cl
```

```python
# Tính số lượng của mỗi mặt hàng
item_count = cl.Counter(chipo['item_name'])

# Sắp xếp các mặt hàng theo số lượng bán được và lấy ra 5 mặt hàng đầu tiên
top_items = item_count.most_common(5)

# Tạo histogram của các mặt hàng này
labels = [item[0] for item in top_items]
values = [item[1] for item in top_items]

plt.figure(figsize = (15, 5))
plt.bar(labels, values, color='blue')
plt.title('Top 5 Items Bought')
plt.xlabel('Item Name')
plt.ylabel('Number of Sales')

# Show the histogram
plt.show()
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
plt.figure(figsize = (20, 10))
plt.scatter(x = chipo['item_price'], y = chipo['order_id'])

# Show the histogram
plt.show()
```

### Step 7. BONUS: Create a question and a graph to answer your own question.

```python
# Tính tổng số lượng mặt hàng được mua cho mỗi loại hàng
item_quantity = chipo.groupby('item_name')['quantity'].sum()

# Vẽ biểu đồ đường
plt.figure(figsize=(20,5))
plt.plot(item_quantity.index, item_quantity.values, color='blue')
plt.title('Total Items Ordered for Each Item')
plt.xlabel('Items')
plt.ylabel('Total Quantity Ordered')

# Hiển thị biểu đồ đường
plt.show()
```
